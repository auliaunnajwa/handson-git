```python
df = df[numbers]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
X_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) #Mengambil 2 komponen utama
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
pcs = pca.transform(X_scaled)

df_pca = pd.DataFrame(data = pcs, columns = ['PC 1', 'PC 2'])
df_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt 

pca_full = PCA()
pca_full.fit(X_scaled)
expl_var = pca_full.explained_variance_ratio_
plt.figure(figsize=(15, 5))
plt.plot(range(1, len(expl_var)+1), expl_var, marker='o')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()
```

```python
df_scaled = pd.DataFrame(X_scaled, columns = df.columns)
df_scaled.head()
```

```python
from sklearn.cluster import KMeans
inertia = []

kmeans = KMeans(n_clusters=5, random_state=42, n_init=10).fit(df_scaled)
clusters = kmeans.labels_
df['clusters'] = clusters
df.head(10)
```

```python
df_pca['clusters'] = clusters

fig, ax = plt.subplots(figsize=(15,5))

sns.scatterplot(
    x='PC 1', y='PC 2',  
    hue='clusters',
    linestyle='--',
    data=df_pca,
    s=100,
    ax=ax
)

plt.show()
```

```python
df.groupby('clusters').median()
```

```python
df['clusters'].value_counts().to_frame().reset_index().rename(columns={"index": "clusters", "clusters": "total_members"})
```

```python
from sklearn.metrics import silhouette_score
#Hitung Silhouettes Score
sil_score = silhouette_score(X_scaled, clusters)

print("Silhouette Score:", sil_score)
```

```python
# List untuk menyimpan hasil 
k_values = []
silhouette_values = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters = k, n_init=10, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)

    k_values.append(k)
    silhouette_values.append(score)
    
# Membuat tabel
df_scores = pd.DataFrame({
    "k": k_values,
    "Silhouette Score" : silhouette_values
})

df_scores
```
