```python
#Konversi kolom waktu
df['upload_date'] = pd.to_datetime(df['upload_date'], errors='coerce')

df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
#IQR
Q1 = df[numerik].quantile(0.25)
Q3 = df[numerik].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Filter baris outliner
df_outlier = df[((df[numerik] < lower_bound) | (df[numerik] > upper_bound)).any(axis=1)]
print('Data outlier: ', df_outlier.shape[0], 'baris')
```

## Feature Engineering

```python
df['year'] = df['upload_date'].dt.year
df['month'] = df['upload_date'].dt.month
df['day_of_week'] = df['upload_date'].dt.day_name()
df['hour'] = df['upload_date'].dt.hour

df.head()
```

```python
# Menambah kolom engagement rate
df['engagement_rate'] = (df['likes'] + df['comments'] + df['shares']) / df['impressions']

df.head()
```

```python
df.to_csv("youtube_cleaned.csv", index=False)
print("File: youtube_cleaned.csv")
```
